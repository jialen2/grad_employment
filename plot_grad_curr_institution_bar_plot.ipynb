```python
import pandas as pd
from matplotlib import pyplot as plt
import os
def cm_to_inch(value):
    return value/2.54
def grad_curr_institution_bar_plot(major, institution):
    major = major.replace(" ", "_")
    institution = institution.replace(" ", "_")
    # Read CSV into pandas
    path = os.getcwd()+"/grad_current_institution/"+institution+"/"+major+".csv"
    data = pd.read_csv(path, header=None)
    data.head()
    df = pd.DataFrame(data)

    name = df.iloc[:,0]
    num_people = df.iloc[:,1]

    # Figure Size
    fig, ax = plt.subplots(figsize =(50, 30))

    # Horizontal Bar Plot
    ax.barh(name, num_people)

    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)

    # Remove x, y Ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')

    # Add padding between axes and labels
    ax.xaxis.set_tick_params(pad = 5)
    ax.yaxis.set_tick_params(pad = 10)

    # Add x, y gridlines
    ax.grid(b = True, color ='grey',
            linestyle ='-.', linewidth = 0.5,
            alpha = 0.2)

    # Show top values
    ax.invert_yaxis()
    # Add annotation to bars
    for i in ax.patches:
        plt.text(i.get_width()+0.2, i.get_y()+0.5,
                 str(round((i.get_width()), 2)),
                 fontsize = 10, fontweight ='bold',
                 color ='grey')
    # Add Plot Title
    ax.set_title('Sports car and their price in crore',
                 loc ='left', )
    
    # Show Plot
    plt.show()
    
grad_curr_institution_bar_plot("combined", "Heritage Institute of Technology")
```
